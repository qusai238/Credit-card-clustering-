```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

# Data Preprocessing

```python
# dropping the cust_ID column
data.drop('CUST_ID',axis=1,inplace=True)
```

```python
#dropping the single null value in the credit_limit 
data.dropna(subset=['CREDIT_LIMIT'],inplace=True)
```

```python
#replacing the minimum_payments null values with the median 
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].median(),inplace=True)
```

```python
#checkinh on null values again, here we see that we have no null values after dealing with them 
data.isnull().sum()
```

# Data Visualization

```python
plt.figure(figsize=(20,35))
for i, col in enumerate(data.columns):
    if data[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sns.kdeplot(data[col], ax=ax)
        plt.xlabel(col)
        
#from the visualization below we can notice that we have a great amount of skewness in our data
#we are going to try and fix that 
```

```python
cols = ['BALANCE', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'ONEOFF_PURCHASES_FREQUENCY',
        'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 
        'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT']
```

```python
for col in cols:
    data[col] = np.log(1 + data[col])
```

```python
plt.figure(figsize=(15,20))
for i, col in enumerate(cols):
    ax = plt.subplot(6, 2, i+1)
    sns.kdeplot(data[col], ax=ax)
```

```python
#Trying to find some correlation using heatmaps
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(),annot=True)
# as you can notice we have some correlated features 
```

# Dimentionality Reductionn using PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
reduced_data=pca.fit_transform(data)
```

```python
reduced_data.shape 
```

# Model Training

```python
#using the elbow method to find the optimal no.of clusters k
# where WCSS stands for within clusters summation of squares which is equal to the inertia
from sklearn.cluster import KMeans
kmeans_models= [KMeans(n_clusters=k,random_state=23).fit(reduced_data) for k in range(1,10)]
models_inertia= [model.inertia_ for model in kmeans_models]
plt.plot(range(1, 10), models_inertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
# as we can see the inertia starts to decrease with a lower rate at 3 or 4 clusters (elbow point)
```

```python
#now lets compare the silhouette score to make sure we pick the optimal k (no.of clusters)
from sklearn.metrics import silhouette_score

silhoutte_scores = [silhouette_score(reduced_data, model.labels_) for model in kmeans_models[1:4]]
plt.plot(range(2,5), silhoutte_scores, "bo-")
plt.xticks([2, 3, 4])
plt.title('Silhoutte scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutte score')
#it appears that k=2 gives the highest silhouette score so we will use k=2 
```

```python
kmeans=KMeans(n_clusters=2,random_state=23)
kmeans.fit(reduced_data)

print('Silhoutte score of our model is ' + str(silhouette_score(reduced_data, kmeans.labels_)))
```

```python
#assigning the labels as cluster index to our data set 
data['CLUSTER_ID']= kmeans.labels_
data
```

```python
#Inverse transforming the log transformation that we did earlier to visualize the results on original scale.
for col in cols:
    data[col] = np.exp(data[col])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='ONEOFF_PURCHASES', y='PURCHASES', hue='CLUSTER_ID')
plt.title('Distribution of clusters based on One off purchases and total purchases')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='CREDIT_LIMIT', y='PURCHASES', hue='CLUSTER_ID')
plt.title('Distribution of clusters based on Credit limit and total purchases')
```

```python
#looking at the above plots we can notice that our model clustered the customers with low usage of credit cards in one side 
#and customers with high usage of credit cards on the other side 
```
